```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['Sleep Duration']=le.fit_transform(df['Sleep Duration'])
df['Dietary Habits']=le.fit_transform(df['Dietary Habits'])
df['Have you ever had suicidal thoughts ?']=le.fit_transform(df['Have you ever had suicidal thoughts ?'])
df['Family History of Mental Illness']=le.fit_transform(df['Family History of Mental Illness'])
df['Depression']=le.fit_transform(df['Depression'])
```

```python
df.head()
```

```python
x=df.drop(['Depression','index'],axis=1)
y=df['Depression']
```

```python
for col in x.columns:
    Q1 = x[col].quantile(0.25)
    Q3 = x[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    x[col] = np.where(x[col] < lower, lower,
                      np.where(x[col] > upper, upper, x[col]))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train_norm=scaler.fit_transform(x_train)
x_test_norm=scaler.fit_transform(x_test)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_resampled, y_resampled = smote.fit_resample(x_train_norm, y_train)
```

```python
import torch
from torch import nn # nn contains all of PyTorch's building blocks for neural networks
import matplotlib.pyplot as plt
```

```python
x_train = torch.tensor(x_resampled, dtype=torch.float32)
x_test = torch.tensor(x_test_norm, dtype=torch.float32)
y_train = torch.tensor(y_resampled, dtype=torch.float32).unsqueeze(1)
y_test = torch.tensor(y_test, dtype=torch.float32).unsqueeze(1)
```

```python
class DP(nn.Module):
  def __init__(self,input_dim):
    super().__init__()
    self.model=nn.Sequential(
        nn.Linear(input_dim,128 ),
        nn.ReLU(),
        nn.Linear(128,32),
        nn.Dropout(0.3),
        nn.ReLU(),
        nn.Linear(32,1),
        nn.Sigmoid()
    )
  def forward(self,x):
    return self.model(x)
```

```python
model = DP(input_dim=x_train.shape[1])
loss=nn.BCEWithLogitsLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=0.001)
```

```python
epochs=100

for epoch in range(epochs):
  model.train()
  y_pred=model(x_train)
  train_loss=loss(y_pred,y_train)
  optimizer.zero_grad()
  train_loss.backward()
  optimizer.step()
  model.eval()

  with torch.inference_mode():
    test_pred=model(x_test)
    test_loss=loss(test_pred,y_test)

    if epoch % 10==0:
      display(f"Epoch: {epoch} | Train loss: {train_loss} | Test loss: {test_loss}")
```

```python
model.eval()
with torch.inference_mode():
    y_pred_prob = model(x_test)
    y_pred_class = torch.round(y_pred_prob) # Convert probabilities to binary predictions

accuracy = accuracy_score(y_test.numpy(), y_pred_class.numpy())
display(f"Accuracy of the PyTorch model on the test set: {accuracy:.4f}")
```

```python
class MLP(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.model(x)
```

```python
model = MLP(input_dim=x_train.shape[1])
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 150
for epoch in range(epochs):
    model.train()
    y_pred = model(x_train)
    loss = criterion(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}/{epochs} - Loss: {loss.item():.4f}")

# Evaluate
model.eval()
with torch.no_grad():
    y_pred1 = model(x_test)
    predicted_classes = (y_pred1 > 0.5).float()
    accuracy = (predicted_classes == y_test).float().mean()
    print(f"\nTest Accuracy: {accuracy:.4f}")
```
